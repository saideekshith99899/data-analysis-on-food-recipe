```python
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
sodium_bins = pd.cut(data['sodium'], bins=[0, 100, 200, 300, 400, 500, 600], labels=['0-100', '101-200', '201-300', '301-400', '401-500', '501-600'])
sodium_counts = sodium_bins.value_counts()
plt.pie(sodium_counts, labels=sodium_counts.index, autopct='%1.1f%%')
plt.title('Sodium Distribution')
plt.show()
```

```python
sns.histplot(data['rating'], bins=10, kde = True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#scatter plot
plt.scatter(data['calories'], data['protein'], alpha=1)
plt.title('Calories vs Protein')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

```python
numeric_columns = ['rating', 'calories', 'protein', 'fat', 'sodium']
corr_matrix = data[numeric_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap = 'rainbow')
plt.show()
```

```python
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.show()
```
